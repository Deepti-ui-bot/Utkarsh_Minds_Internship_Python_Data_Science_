```python
import pandas
import numpy
import matplotlib.pyplot as plt
```

```python
data = pandas.read_csv('/content/StudentStatistics.csv', sep=';')
data.head()
```

```python
selected_data = data[['Teach_Motivate', 'Teach_LinkTheory', 'Teach_StimAsk', 'Teach_Avail']]
selected_data
```

```python
selected_data['Teach_Motivate'].unique()
```

```python
coding = {'Fully Disagree':0, 'Disagree':1, 'Neither disagree nor agree':2, 'Agree':3, 'Fully agree':4}
selected_data.replace(coding, inplace=True)
```

```python
selected_data.dropna(inplace=True)
```

```python
from scipy.stats import friedmanchisquare
#'Teach_Motivate', 'Teach_LinkTheory', 'Teach_StimAsk', 'Teach_Avail'

friedmanchisquare(selected_data['Teach_Motivate'], selected_data['Teach_LinkTheory'], selected_data['Teach_StimAsk'], selected_data['Teach_Avail'])
```

```python
statistic=33.85869565217378
pvalue=2.1220432262951938e-07
```

```python
#Effect size - Kendall W

n = selected_data.shape[0]
k = selected_data.shape[1]

W = statistic / (n * (k-1))
W
```

image.png

```python
0 - 0.3 - weak
0.3 - 0.5 - moderate
0.5 - 0.7 -relatively strong
0.7+ - strong
```

The difference were not having the same responses and after

applying the Friedman test the difference was significant (p<0.05)

and further the Kendal W effect size suggest the difference is weak (W = 0.21)
