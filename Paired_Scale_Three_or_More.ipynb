```python
import numpy
import pandas
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
X, y =  load_iris(return_X_y = True, as_frame = True)
X.head()
```

```python
y.head()
```

```python
y.unique()
```

There are 3 flowers numbered as 0, 1,and 2.

```python
data = pandas.concat([X,y], axis = 1)   #concat by column X and y
```

We need X and y together, so we will concat them with columns ( axis=1).

```python
data
```

```python
data['target'] = data['target'].replace({0:'type1', 1:'type2', 2:'type3'})
```

```python
data.columns
```

These colunm names have space in between and brackets also, which we must remove, because it may create problem in analysis using MANOVA function; so rename them

```python
#MANOVA
from statsmodels.multivariate.manova import MANOVA

data = data.rename(columns = {
    'sepal length (cm)':'sepal_length', 'sepal width (cm)':'sepal_width', 'petal length (cm)':'petal_length',
       'petal width (cm)':'petal_width', 'target':'flower_name'
})

data.columns
```

```python
result = MANOVA.from_formula('sepal_length + sepal_width + petal_length + petal_width ~ flower_name',data=data)
```

```python
print(result.mv_test())
```

 Wilks' lambda is the effect size

```python
Wilks_lambda = 0.0234 #Effect size

pvalue = 0.0000
```

```python
> 0.9     #if greater than 0.9 then small
small

> 0.7     #if greater than 0.7 than meium but small than 0.9
medium

else      #else if less than 0.7 than large difference
large
```

MANOVA test found that sepal_length, sepal_width, petal_length, petal_width are

not same for different flower_name (pvalue<0.05)

Further Wilks lambda effect size suggest that the difference is large (Wilks_lambda = 0.0234)
