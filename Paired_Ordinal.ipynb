```python
import numpy
import pandas

num_patients = 20

before = numpy.random.randint(1,5, size = num_patients)
after = numpy.random.randint(1,5, size = num_patients)

data = pandas.DataFrame(
    {
        'before': before,
        'after': after
    }
)

data
```

```python
crosstab = pandas.crosstab(data['before'], data['after'])
crosstab
```

```python
#bar chart
crosstab.plot(kind='bar')
```

```python
from scipy.stats import wilcoxon

wilcoxon(data['before'], data['after'], correction=False)
```

```python
statistic=51.0
pvalue=0.5959628483529898
```

```python
from scipy.stats import norm

# Rosenthal correlation coefficient - effect size
z_value = norm.ppf(1 - (pvalue/2))

import math

r_ros = abs(z_value / math.sqrt(data.shape[0]))

r_ros
```

```python
if r_ros < 0.20:
  print('very low')
elif r_ros < 0.40:
  print('low')
elif r_ros < 0.60:
  print('moderate')
elif r_ros < 0.80:
  print('strong')
else:
  print('very strong')
```

```python
#The effect size is very low and effect of the event on the patients as we applied the wilcoxon test
```
