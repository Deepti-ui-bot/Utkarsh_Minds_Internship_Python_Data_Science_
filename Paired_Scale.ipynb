```python
import numpy
import pandas
import matplotlib.pyplot as plt

num_individual = 50

before = numpy.random.normal(loc = 70, scale = 10, size = num_individual)
after = numpy.random.normal(loc = 30, scale = 20, size = num_individual)

data =  pandas.DataFrame(
    {
        'before': before,
        'after': after
    }
)

data
```

```python
data.describe()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(8,8))

ax[0].hist(data['before'], bins=10, color = 'blue', edgecolor = 'black')
ax[1].hist(data['after'], bins=10, color = 'red', edgecolor = 'black')
plt.tight_layout
plt.show()
```

```python
from scipy.stats import ttest_rel

ttest_rel(data['before'], data['after'])
```
