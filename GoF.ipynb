```python
import pandas
data = pandas.read_csv('GSS2012a.csv')
data
```

```python
data['mar1'].value_counts()
```

RQ - Are people more likely to get married?

H0 - there is no preference and the distribution of each category is same

Ha - there is no preference and the distribution of each category is not same

```python
from scipy.stats import chisquare
chisquare(data['mar1'].value_counts())
```

```python
statistic=600.7191011235955
pvalue=1.0829030167514616e-128
```

```python
# Cramer V
# n = samples
# df = number of categories -1

n = data.shape[0]
df = 4
cv = (statistic / (n * df)) ** 0.5
cv
```

```python
In the population the different categories are not evenly distributed
as per the Pearson chisquare goodness-of-fit test (p<0.05) and the
difference in the categories can be calculated from below range (given thatCramer V =0.408).
```
