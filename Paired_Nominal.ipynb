```python
import pandas
import numpy

categories = ['cat', 'dog', 'parrot', 'snake']

num_individual =50

before = numpy.random.choice(categories, size = num_individual)

after = numpy.random.choice(categories, size = num_individual)
```

```python
data = pandas.DataFrame({
    'before': before,
    'after': after
})
data
```

```python
crosstable = pandas.crosstab(data['before'], data['after'])
crosstable
```

```python
crosstable.plot(kind='bar')
```

```python
from statsmodels.stats.api import SquareTable

print(SquareTable(crosstable, shift_zeros=False).homogeneity(method='bhapkar'))
```

```python
pvalue < 0.05; H0 rejected
no difference
```
